```python
df_building_subset = df_lbsm[['postcode_locator', 'construction_age_band', 'conservation_area_flag']].copy()

# 2. Drop duplicates based on the postcode
# This ensures each postcode has only one entry (the first one found)
df_building_subset = df_building_subset.drop_duplicates(subset=['postcode_locator'])

# 3. Rename the column to match your price dataset for an easy merge
df_building_subset = df_building_subset.rename(columns={'postcode_locator': 'postcode'})
```
